```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prod_df = pd.read_table('./data/products.tsv')
rev_df = pd.read_table('./data/reviews.tsv')
prod_df.head()
```

```python
# Most popular brand
z = rev_df.productId.value_counts()
z[z.index[0]]
```

```python
prod_name = list()
popularity = list()
for i in range(0,len(z.index)):
    prod_name.append(prod_df.iloc[prod_df[prod_df.product_id == z.index[i]].index[0]].product_name+" "
    +prod_df.iloc[prod_df[prod_df.product_id == z.index[i]].index[0]].product_model)
    popularity.append(z[z.index[i]])
```

```python
plt.figure(facecolor='#f5f2eeff')
plt.axes().set_facecolor("#f5f2eeff")

plt.barh(prod_name,popularity,color='#9d6cab')
plt.show()
```
